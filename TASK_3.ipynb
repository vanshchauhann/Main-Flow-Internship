```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('//content//householdtask3.csv')#read
```

```python
display(data.head(10))#print 10 rows of the data set
```

```python
#Scatter plot with year column against own
plt.scatter(data['year'], data['own'])

#Adding title to the plot
plt.title('Year vs Own')

#Adding labels to the x-axis and y-axis
plt.xlabel('Year')
plt.ylabel('Own')

#Displaying the plot
plt.show()
```

```python
#Line chart with year against own
plt.plot(data['year'])
plt.plot(data['own'])

#Adding title to the plot
plt.title('Line Chart')

#Adding labels to the x-axis and y-axis
plt.xlabel('Year')
plt.ylabel('Own')

#display
plt.show()
```

```python
#Bar chart or bar plot
plt.bar(data['year'], data['own'])

#Adding title to the plot
plt.title('Bar Chart')

#Adding labels to the x-axis and y-axis
plt.xlabel('Year')
plt.ylabel('Own')

#display
plt.show()
```

```python
#Histogram
plt.hist(data['income'])

#Adding title to the plot
plt.title('Histogram')

#display
plt.show()
```
